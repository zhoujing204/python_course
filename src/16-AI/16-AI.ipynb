神经元输出的计算公式

- `b`被称为`bias`（偏置），用于控制神经元被激活的容易程度
- `w`被称为`weight`（权重），用于控制输入信号的权重
- `h()`被称作`activation function`（激活函数），用于控制神经元的输出

$a = b + w_1x_1 + w_2x_2$

$y = h(a)$

常见的激活函数：

- sigmoid（早期）
- ReLU（现在）

sigmoid函数的公式：

$h(x) = \frac{1}{1 + e^{-x}}$

```python
# sigmoid函数的实现
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# sigmoid函数的可视化

import matplotlib.pyplot as plt
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) # 指定y轴的范围
plt.show()
```

ReLU函数的公式：

$h(x) = \begin{cases} x & (x > 0) \\ 0 & (x \leq 0) \end{cases}$

```python
# ReLU函数的实现
import numpy as np

def relu(x):
    return np.maximum(0, x)
```

```python
# ReLU函数的可视化
import matplotlib.pyplot as plt
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-1.0, 5.5) # 指定y轴的范围
plt.show()
```
